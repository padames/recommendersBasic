```python
import pandas as pd
```

```python
ratings = pd.read_csv('data/ratings.csv')
movies = pd.read_csv('data/movies.csv')
#merge the datasets
ratings = pd.merge(movies,ratings).drop(['genres','timestamp'],axis=1)
```

```python
#### Since we already have a ratings feature, we can drop some columns
ratings.head()
```

```python
#pivot to get it to the same format as the previous dataframe
```

```python
user_ratings = ratings.pivot_table(index=['userId'],columns=['title'],values='rating')
```

```python
user_ratings.head()
```

```python
#Let's remove movies that have less than 10 users who have rated the movie
```

```python
user_ratings = user_ratings.dropna(thresh=10,axis=1).fillna(0)
```

```python
user_ratings
```

```python
user_ratings.describe()
```

```python
#Let's build our similarity matrix but using pearson correlation instead 
# of cosine similarity
```

```python
#This method adjusts for the means.
#Hence, we do not need to adjust to the zeroes we filled in for the nan values
item_similarity_df = user_ratings.corr(method='pearson')
```

```python
item_similarity_df
```

```python
def get_similar_movies(movie_name,user_rating):
    #we are gettting the similar score of that movie
    # and scaling the row of the movie by the user rating
    #we subtract the user_rating by the mean which is 2.5 to correct for the error
    #seen in the above cell
    similar_score = item_similarity_df[movie_name]*(user_rating-2.5)
    similar_score = similar_score.sort_values(ascending=False)
    return similar_score
```

```python
get_similar_movies("10,000 BC (2008)",5)
```
